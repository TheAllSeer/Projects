```python
import re
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
```

```python
url = 'https://www.apartments.com/orlando-fl/'  # apartments in orlando? apartments.com anyways
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
time.sleep(.5)
driver.get(url)

listings = driver.find_elements(By.CSS_SELECTOR, 'div#placardContainer>ul>li')
l_dict = {}

for i in range(len(listings)):
    l = listings[i]
    dd = {}
    try:
        title = l.find_element(By.CLASS_NAME, 'property-title')
        dd['title'] = title.text
    except NoSuchElementException:
        dd['title'] = "no"
    try:
        address = l.find_element(By.CLASS_NAME, 'property-address')
        dd['address'] = address.text
    except NoSuchElementException: 
        dd['address'] = "no"
    try:
        company = l.find_element(By.CLASS_NAME, 'property-logo').get_attribute("aria-label")
        dd['company'] = company
    except NoSuchElementException:
        dd['company'] = 'no'
    try:
        pricing = l.find_element(By.CLASS_NAME, 'property-pricing')
        dd['pricing'] = pricing.text
    except NoSuchElementException:
        dd['pricing'] = 'no'
    try:
        beds = l.find_element(By.CLASS_NAME, 'property-beds')
        dd['beds'] = beds.text
    except NoSuchElementException: 
        dd['beds'] = 'no'
        continue
    try:
        specials = l.find_element(By.CLASS_NAME, 'property-specials')
        dd['specials'] = specials.text
    except NoSuchElementException:  
        dd['specials'] = 'no'
    try:
        link = l.find_element(By.CLASS_NAME, 'property-link').get_attribute('href')
        dd['link'] = link
    except NoSuchElementException:  
        dd['link'] = 'no'
    try:
        amenities = l.find_elements(By.CSS_SELECTOR, '.property-amenities>span')
    except NoSuchElementException:
        dd['amenities'] = 'no'
  
    amen = [a.text for a in amenities]
    dd['amenities'] = amen
    try:
        actions = l.find_element(By.CLASS_NAME, 'property-actions')
    except NoSuchElementException:  
        continue
 
    try:
        phone = actions.find_element(By.CLASS_NAME, 'phone-link')
        dd['phone'] = 'phone.text (redacted for privacy reasons)'
    except NoSuchElementException:
        dd['phone'] = 'no'

    try:
        mail = actions.find_element(By.CLASS_NAME, 'checkAvailability')
        dd['mail'] = 'mail.text (redacted for privacy reasons)'
    except NoSuchElementException:
        dd['mail'] = 'no'
    l_dict['{}'.format(i)] = dd



df = pd.DataFrame(l_dict).transpose()
df
```
