```python
import re
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
```

```python
url = 'https://www.yad2.co.il/realestate/rent?topArea=25&area=5&city=4000'
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
time.sleep(1)
driver.get(url)
```

```python
listings = driver.find_elements(By.CLASS_NAME, 'feeditem')
print("price: {}".format(listings[0].find_element(By.CLASS_NAME, 'price').text))
print("title: {}".format(listings[0].find_element(By.CLASS_NAME, 'title').text))
print("renter: {}".format(listings[0].find_element(By.CLASS_NAME, 'merchant_name').text))
print("rooms: {}".format(listings[0].find_element(By.CLASS_NAME, 'rooms-item').text))
print("floor: {}".format(listings[0].find_element(By.CLASS_NAME, 'floor-item').text))
print("SQ meters: {}".format(listings[0].find_element(By.CLASS_NAME, 'SquareMeter-item').text))
print("img link: {}".format(listings[0].find_element(By.CLASS_NAME, 'feedImage').get_attribute('src')))
```

```python
l_dict = {}
for i in range(len(listings)):
    dd = {}
    try:
        price = listings[i].find_element(By.CLASS_NAME, 'price').text
    except NoSuchElementException:
        price = 'no'
    try:
        title = listings[i].find_element(By.CLASS_NAME, 'title').text
    except NoSuchElementException:
        title = 'no'
    try:
        subtitle = listings[i].find_element(By.CLASS_NAME, 'subtitle').text
    except NoSuchElementException:
        subtitle = 'no'
    try:
        merchant = listings[i].find_element(By.CLASS_NAME, 'merchant_name').text
    except NoSuchElementException:
        merchant = 'no'
    try:
        rooms = listings[i].find_element(By.CLASS_NAME, 'rooms-item').text
    except NoSuchElementException:
        rooms = 'no'
    try:
        floor = listings[i].find_element(By.CLASS_NAME, 'floor-item').text
    except NoSuchElementException:
        floor = 'no'
    try:
        squareMeter = listings[i].find_element(By.CLASS_NAME, 'SquareMeter-item').text
    except NoSuchElementException:
        squareMeter = 'no'
    try:
        imageLink = listings[i].find_element(By.CLASS_NAME, 'feedImage').get_attribute('src')
    except NoSuchElementException:
        imageLink = 'no'
    dd['title'] = title
    dd['subtitle'] = subtitle
    dd['price'] = price
    dd['merchant'] = merchant
    dd['rooms'] = rooms
    dd['floor'] = floor
    dd['squareMeter'] = squareMeter
    dd['imageLink'] = imageLink
    l_dict['{}'.format(i)] = dd
driver.quit()
```

```python
df = pd.DataFrame(l_dict).transpose()
df
```

```python
df['rooms'] = df['rooms'].str.replace("\nחדרים", ' rooms')
df['floor'] = df['floor'].str.replace("\nקומה", ' floor')
df['floor'] = df['floor'].str.replace("קרקע", 'ground')
df['squareMeter'] = df['squareMeter'].str.replace("""\nמ"ר""", ' squared meters')
df
```
