```python
import re
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import WebDriverException
```

```python
url = 'https://www.upwork.com/search/profiles/?q=web%20scraping'
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
time.sleep(1)
driver.get(url)
```

```python
commonwords = pd.read_csv("d:/programming/python/commonWordsUpwork.csv")['0'].tolist()

clamped_texts = driver.find_elements(By.CLASS_NAME, 'clamped')
words_dict = {}
for ct in clamped_texts:
    ct_split = ct.text.split()
    for w in ct_split:
        word = w.lower().replace(',', '').replace('.', '').replace(':', '')
        words_dict[word] = words_dict.get(word, 0) + 1
words_dict

```

```python
words_by_use = pd.DataFrame(pd.Series(words_dict)).rename(columns = {pd.DataFrame(pd.Series(words_dict)).columns[0]:'count'}).sort_values(by = 'count', ascending = False)
```

```python
df = words_by_use.copy()
df.iloc[0].name
df
```

```python
common_words = []
done = []
```

```python
dfB4 = df[df['count'] >= df.iloc[0]/10]
for i in range(len(df[df['count'] > 4])):
    word = dfB4.iloc[i].name
    if word in done:
        continue
    ans = input("is the word [ {} ] too general? (y/n, cancel to break.)".format(word))
    done.append(word)  # appends the words I've already answered to a list so i dont have to do them again later
    if ans == 'y':
        common_words.append(word)
    if ans == 'cancel':
        break
    else:
        continue

```
